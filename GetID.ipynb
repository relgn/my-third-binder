```python
import pandas as pd

iDent_Input = input("Enter NRIC/FIN:")
iDent_List = iDent_Input.split()



for i in range(len(iDent_List)):
    iDent = iDent_List[i]   

    if len(str(iDent)) == 9:
            
        LLS = iDent[0:1]
        Num1 = int(iDent[1:2])
        Num2 = int(iDent[2:3])
        Num3 = int(iDent[3:4])    
        Num4 = int(iDent[4:5])    
        Num5 = int(iDent[5:6])    
        Num6 = int(iDent[6:7])    
        Num7 = int(iDent[7:8])
        LLE = iDent[8:9]   

        Sum = (2 * Num1) + (7 * Num2) + (6 * Num3) + (5 * Num4) + (4 * Num5) + (3 * Num6) + (2 * Num7)

        
        if LLS == "S" or LLS == "F":
            Rem_Num = Sum % 11
            

        elif LLS == "G" or LLS == "T":
            Rem_Num = (Sum + 4) % 11

        elif LLS == "M":
            Rem_Num = (Sum + 3) % 11
        

        if LLS == "S" or LLS == "T":
            if Rem_Num == 0:
                LLC = "J"
            
            elif Rem_Num == 1:
                LLC = "Z"

            elif Rem_Num == 2:
                LLC = "I"
            
            elif Rem_Num == 3: 
                LLC = "H"
            
            elif Rem_Num == 4: 
                LLC = "G"
            
            elif Rem_Num == 5: 
                LLC = "F"
            
            elif Rem_Num == 6: 
                LLC = "E"
            
            elif Rem_Num == 7: 
                LLC = "D"
            
            elif Rem_Num == 8: 
                LLC = "C"
            
            elif Rem_Num == 9: 
                LLC = "B"
            
            elif Rem_Num == 10: 
                LLC = "A"
        
        else:
            if Rem_Num == 0:
                LLC = "X"
            
            elif Rem_Num == 1:
                LLC = "W"

            elif Rem_Num == 2:
                LLC = "U"
            
            elif Rem_Num == 3: 
                LLC = "T"
            
            elif Rem_Num == 4: 
                LLC = "R"
            
            elif Rem_Num == 5: 
                LLC = "Q"
            
            elif Rem_Num == 6: 
                LLC = "P"
            
            elif Rem_Num == 7: 
                LLC = "N"
            
            elif Rem_Num == 8: 
                LLC = "M"
            
            elif Rem_Num == 9: 
                LLC = "L"
            
            elif Rem_Num == 10: 
                LLC = "K"

        if LLE == LLC:
            print(iDent + ", Check-sum passed, ID is valid")  
        else:
            print(iDent + ", Check-sum failed, '" + LLC + "' should be the Ending letter")      
    else:
        print("ID is invalid")        
```
